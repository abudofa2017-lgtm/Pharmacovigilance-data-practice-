```python
#importing necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

**Loading the Data**

```python
# Loading the dataet
iris_df = pd.read_csv('iris.csv', header=None,
                      names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])

# Displayin first few rows
print(iris_df.head())
```

**Preparing samples array(features only)**

```python
# Extracting feature columns as a NumPy array (samples)
samples = iris_df.iloc[:, :4].values   # first 4 columns: measurements
print("Samples shape:", samples.shape)
```

**Creating K-Means model with 3 clusters and fitting the model**

```python
# Creating KMeans instance
model = KMeans(n_clusters=3, random_state=42)

# Fitting the model to the samples
model.fit(samples)

# Getting cluster labels for each sample
labels = model.predict(samples)

print("First 10 cluster labels:", labels[:10])
```

**Visualising clustering (sepal length vs petal  length)**

```python
# x = sepal length (column 0), y = petal length (column 2)
xs = samples[:, 0]   # sepal length
ys = samples[:, 2]   # petal length

plt.figure(figsize=(8, 5))
plt.scatter(xs, ys, c=labels, cmap='viridis', alpha=0.7, edgecolors='k')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.title('K-Means Clustering of Iris (3 clusters)')
plt.colorbar(label='Cluster')
plt.show()
```

**Predicting clusters for new samples**

```python
# New samples: each row = [sepal_length, sepal_width, petal_length, petal_width]
new_samples = np.array([[5.0, 3.2, 1.5, 0.2],   # looks like setosa
                        [6.5, 3.0, 5.2, 2.0]])  # looks like virginica

new_labels = model.predict(new_samples)
print("Cluster labels for new samples:", new_labels)
```

**Interpretation of the Plot**

```python
"""
The scatter plot shows three distinct clusters (colours) when plotting sepal length vs petal length.
- Cluster 0 (left-bottom): low sepal and low petal length → corresponds to Iris-setosa.
- Cluster 1 (middle): medium values → Iris-versicolor.
- Cluster 2 (right-top): high sepal and high petal length → Iris-virginica.
K-Means successfully separates the three species using only the measurement features.
"""
```
